```python
import pandas as pd

# Load sample of delays, different train stations 
archivo_excel = "sample_delays.xlsx"

# reading all pages 
hojas = pd.read_excel(archivo_excel, sheet_name=None)

#iterating through each sheet and concatenating dataframes as well as adding new column with estation name
dataframes = []
for nombre_hoja, df in hojas.items():
    df["Estación"] = nombre_hoja  
    dataframes.append(df)


df_final = pd.concat(dataframes, ignore_index=True)


display(df_final)
```

```python
# Renaming columns
df_final.rename(columns={
    "Hora de llegada": "arrival_time",
    "Origen": "origin",
    "Tren": "train",
    "FECHA": "date",
    "Estación": "train_station"
}, inplace=True)

#delegint rows with values NaN in arrival_time, origin, train
df_final.dropna(subset=["arrival_time", "origin", "train"], inplace=True)


display(df_final)
```

```python
# Filter the DataFrame to keep only relevant train types
train_keywords = ["AVLO", "AVE", "IRYO", "OUIGO"]
df_filtered = df_final[df_final["train"].str.contains("|".join(train_keywords), na=False, case=False)]

# Display the filtered DataFrame
display(df_filtered)
```

```python
# Lista de compañías que quieres extraer
companies = ["AVE", "AVLO", "IRYO", "OUIGO"]

def extract_company(train):
    for company in companies:
        if company in train:
            return company
    return None  

# Crear la nueva columna aplicando la función
df_filtered["company"] = df_filtered["train"].apply(extract_company)

display(df_filtered.sample(10))
```

```python

display(df_filtered.sample(10))
```

```python
# Función para separar los timestamps
def process_arrival_time(arrival_time):
    arrival_time = str(arrival_time)  # Convertir a cadena por si es datetime.time
    tiempos = arrival_time.split("\n")  # Separar por salto de línea
    return tiempos[0], tiempos[-1]  # Primer timestamp en arrival_time, último en real_arrival_time

# Aplicar la función asegurando que arrival_time es una cadena
df_filtered["arrival_time"], df_filtered["real_arrival_time"] = zip(*df_filtered["arrival_time"].astype(str).apply(process_arrival_time))

# Mostrar los primeros registros para comprobar
display(df_filtered.sample(10))
```

```python
# Crear la nueva columna comparando arrival_time y real_arrival_time
df_filtered["status"] = df_filtered.apply(lambda row: "delayed" if row["arrival_time"] != row["real_arrival_time"] else "on time", axis=1)

# Mostrar los primeros registros para comprobar
display(df_filtered.sample(10))
```

```python
import scipy.stats as stats
# Filtrar por compañías
renfe = df_filtered[df_filtered['company'].isin(["AVE", "AVLO"])]
otras = df_filtered[df_filtered['company'].isin(["IRYO", "OUIGO"])]

# Calcular proporción de puntualidad
p_renfe = (renfe['status'] == "on time").sum() / len(renfe)
p_otras = (otras['status'] == "on time").sum() / len(otras)

# Prueba de proporciones
n_renfe = len(renfe)
n_otras = len(otras)

prop_test = stats.ttest_ind(renfe['status'] == "on time", otras['status'] == "on time")

print(f"RENFE puntualidad: {p_renfe:.2f}")
print(f"Otras compañías puntualidad: {p_otras:.2f}")
print("Prueba de hipótesis:", prop_test)
```

```python
df = df_filtered.copy()
```

```python
# Convertir la columna date a datetime
df["date"] = pd.to_datetime(df["date"], format="%Y-%m-%d")

# Agregar ":00" a los valores de tiempo si no tienen segundos
df["arrival_time"] = df["arrival_time"].apply(lambda x: x if len(x) == 8 else x + ":00")
df["real_arrival_time"] = df["real_arrival_time"].apply(lambda x: x if len(x) == 8 else x + ":00")

# Convertir a datetime 
df["arrival_time"] = pd.to_datetime(df["date"].astype(str) + " " + df["arrival_time"], format="%Y-%m-%d %H:%M:%S", errors="coerce")
df["real_arrival_time"] = pd.to_datetime(df["date"].astype(str) + " " + df["real_arrival_time"], format="%Y-%m-%d %H:%M:%S", errors="coerce")

# Calcular el retraso en minutos
df["delay_minutes"] = (df["real_arrival_time"] - df["arrival_time"]).dt.total_seconds() / 60
```

```python
df.sample(10)
```

```python
# Prueba Chi-cuadrado para ver si la compañía afecta la puntualidad
contingency_table = pd.crosstab(df["company"], df["status"])
chi2_stat, p_chi2, dof, expected = stats.chi2_contingency(contingency_table)

print("📌 Prueba Chi-cuadrado:")
print(f"Chi2 statistic: {chi2_stat:.2f}, p-value: {p_chi2:.5f}")
```

```python
# 2️⃣ Prueba t para comparar tiempos de retraso entre RENFE (AVE + AVLO) vs. otras compañías
renfe_delays = df.loc[df["company"].isin(["AVE", "AVLO"]), "delay_minutes"]
otras_delays = df.loc[df["company"].isin(["IRYO", "OUIGO"]), "delay_minutes"]

t_stat, p_ttest = stats.ttest_ind(renfe_delays, otras_delays, equal_var=False)  # Test de comparación de medias

print("\n📌 T-test for delay comparison")
print(f"t-statistic: {t_stat:.2f}, p-value: {p_ttest:.5f}")

#Since the p-value is less than 0.05, we can conclude that there is a statistically significant difference between the delays of RENFE (AVE + AVLO) 
# and those of IRYO + OUIGO. That is, RENFE's delay times are significantly different from those of the other companies.
```
